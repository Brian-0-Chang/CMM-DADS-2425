```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df_iris = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df_iris

```

```python
df_iris['Type'] = iris['target']
df_iris
```

```python
### Assign name to each type, instead of 0, 1, and 2, with the following names: setosa, versicolor, and virginica.
### why ? --> classification problem

df_iris['Type'] = df_iris['Type'].replace({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})

#dictionary 이용해서 replace
df_iris
```

## Visualization
## Correlation Plot

```python
df_iris[df_iris.columns[:4]].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig,axes = plt.subplots(figsize=(7,7))
sns.heatmap(df_iris.iloc[:,:4].corr(), annot=True, cbar=True, vmin=1, vmax=1, cmap='crest')
axes.tick_params(labelrotation=45)
plt.title('Correlation Matrix') 
plt.show()  
```

### violin plot


```python
fig,axes = plt.subplots(figsize=(7,7))
sns.violinplot(x="Type", y="sepal length (cm)", data=df_iris)
plt.title('Violin plot of sepal length')
plt.show()
```

```python
fig,axes = plt.subplots(figsize=(7,7))  
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='Type', data=df_iris)  

plt.show()  
```
